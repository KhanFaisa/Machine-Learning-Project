#### Step 1 : Read The Train Datasets 

```python
import pandas as pd 
df = pd.read_csv('Cars93 (2).csv', na_values=['', 'NA'], keep_default_na=False)
df.head() 
```

#### Step 2 : Perform Basic Data Quaity Checks 

```python
df.info()
```

```python
m = df.isna().sum()
m
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(keep='first')
```

#### Step 3 : Seprate X and Y (Wieght)

```python
X = df.drop(columns=['id', 'Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python

Y.head()
```

```python

X.isna().sum()
```

#### Step 4 : Perform Preprocesiing On X Data 

```python
# cat and con seperation
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline # automating steps
from sklearn.impute import SimpleImputer # Replacing Missing values
from sklearn.preprocessing import StandardScaler, OneHotEncoder # Preprocessing
from sklearn.compose import ColumnTransformer # To combine cat and con columns
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('OHE', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
X_pre.to_csv('CarsPre.csv', index=False)
```

### Step 5 : Perform Train Test Split

20% - Unseen to Model (Test)

```python
X_pre.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

## Step 6: Model Building

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

#### Step 7: Model Evaluation

```python
# Check R2 score in train
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute Error : {mae:.2f}')
    print(f'R2 Score : {r2:.4f}')
```

```python
evaluate_model(model, xtrain, ytrain)
```

```python
evaluate_model(model, xtest, ytest)
```
